```python
import numpy as np
from matplotlib import pyplot as plt
```

# Alexnet results reproduction 

```python
accuracies_cp_theirs = [0.5498, 0.7853]
accuracies_cp_ours = [0.3946001838235294, 0.6787913602941177]
accuracies_orig = [0.56522, 0.79066]
labels = ["acc@1", "acc@5"]

fig, ax = plt.subplots(figsize=(8, 4))
x = np.arange(len(labels))
bar_width = 0.2
bar1 = ax.bar(x - bar_width, accuracies_cp_ours, bar_width, label="CP Model (This work)")
bar2 = ax.bar(x, accuracies_cp_theirs, bar_width, label="CP Model (Their work)")
bar3 = ax.bar(x + bar_width, accuracies_orig, bar_width, label="Original Model")

ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
ax.grid()
ax.set_title("Accuracy Comparison")
plt.savefig("figures/accuracies_alexnet.png")
plt.show()
```

```python
plt.figure(figsize=(8, 4))

model_sizes = [61100840, 8513084, 8799651]
labels = ["Original Model", "CP Model (Our work)", "CP Model (Their work)"]

plt.bar(labels, model_sizes)
plt.title("Model Size")
plt.ylabel("Number of Parameters")
plt.grid()
plt.xticks(ticks=labels)


plt.tight_layout()
plt.savefig("figures/model_size_inference_time_alexnet.png")
plt.show()
```
