```python
import numpy as np
from matplotlib import pyplot as plt
```

# Alexnet results reproduction 

```python
accuracies_cp_theirs = [0.4049402573529412, 0.6939682904411765]
accuracies_cp_ours = [0.3946001838235294, 0.6787913602941177]
accuracies_orig = [0.56522, 0.79066]
labels = ["acc@1", "acc@5"]

fig, ax = plt.subplots(figsize=(6, 4))
x = np.arange(len(labels))
bar_width = 0.2
bar1 = ax.bar(x - bar_width, accuracies_cp_ours, bar_width, label="CP Model (Our ranks)")
bar2 = ax.bar(x, accuracies_cp_theirs, bar_width, label="CP Model (Their ranks)")
bar3 = ax.bar(x + bar_width, accuracies_orig, bar_width, label="Original Model")

ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
ax.grid()
ax.set_title("Accuracy Comparison")
plt.savefig("figures/accuracies_alexnet.png")
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 4))

model_sizes = [61100840, 8513084, 8799651]
inference_times = [0.025220155715942383, 0.056990623474121094, 0.058960914611816406]
labels = ["Original Model", "CP Model (Our ranks)", "CP Model (Their ranks)"]

ax[0].bar(labels, model_sizes)
ax[0].set_title("Model Size")
ax[0].set_ylabel("Number of Parameters")
ax[0].grid()
ax[0].set_xticklabels(labels)

ax[1].bar(labels, inference_times)
ax[1].set_title("Inference Time")
ax[1].set_ylabel("Seconds")
ax[1].grid()
ax[1].set_xticklabels(labels)

plt.tight_layout()
plt.savefig("figures/model_size_inference_time_alexnet.png")
plt.show()
```
